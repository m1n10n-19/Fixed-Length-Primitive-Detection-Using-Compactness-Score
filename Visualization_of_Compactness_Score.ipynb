```python
import matplotlib.pyplot as plt
```

```python
def anomaly(cs):
    diff = [cs[1] - cs[0], 0]
    for i in range(1, len(cs)):
        if(diff[0] < cs[i] - cs[i-1]):
            diff[0] = cs[i] - cs[i-1]
            diff[1] = i
    print("Anomaly at ", diff[1]+1)
    return diff[1]+1
```

```python
def plot_cs(cs_file_name, limit = 45):
    cs_file = open(cs_file_name, 'r')
    file_content = cs_file.read()
    cs_file.close()
    cs_str = file_content.split(",")
    cs_str.pop()
    cs = [float(i) for i in cs_str]
    n_range = [i for i in range(1, limit)]
    anomaly(cs)
    print("Max cs score obtained at ",cs.index(max(cs)) + 1," = ", max(cs))
    f = plt.figure()
    f.set_figwidth(21)
    f.set_figheight(13)
    plt.plot(n_range, cs)
    plt.xlabel("Pattern Length", fontsize = 25)
    plt.xticks([i for i in range(1, 45)], fontsize = 15)
    plt.yticks(fontsize = 15)
    #plt.ylim((-0.47, -0.37))
    plt.ylabel("Compactness Score", fontsize = 25)
    plt.vlines(n_range, 0, cs, linestyle="dashed")
    cs_file_name = "CSs/cs_entropy_bit_ASCII_128_all.txt"
    cs_file = open(cs_file_name, 'r')
    file_content = cs_file.read()
    cs_file.close()
    cs_str = file_content.split(",")
    cs_str.pop()
    cs = [float(i) for i in cs_str]
    n_range = [i for i in range(1, limit)]
    anomaly(cs)
    print("Max cs score obtained at ",cs.index(max(cs)) + 1," = ", max(cs))
    plt.plot(n_range, cs)
    plt.xlabel("Pattern Length", fontsize = 25)
    plt.xticks([i for i in range(1, 45)], fontsize = 15)
    plt.yticks(fontsize = 15)
    #plt.ylim((-0.47, -0.37))
    plt.ylabel("Compactness Score", fontsize = 25)
    plt.vlines(n_range, 0, cs, linestyle="dashed")
    plt.show()
```

```python
plot_cs("CSs/cs_bit_ASCII_128_all.txt")
```
