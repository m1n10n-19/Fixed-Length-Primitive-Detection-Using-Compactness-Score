```python
import random

def shuffle_words(word):
    lst = list(word)
    random.shuffle(lst)
    shuffled_word = ''.join(lst[:])
    return shuffled_word
```

```python
#generation of text_1
word = ''.join([chr(i) for i in range(65, 91)]) + ''.join([chr(i) for i in range(97, 123)])   
text = word

text += shuffle_words(word)

text_file = open("Corpuses/text_1.txt", "w")
text_file.write(text)
text_file.close()
```
