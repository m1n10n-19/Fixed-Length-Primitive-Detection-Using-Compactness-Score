```python
def tobinary(ch):
    bit = []
    if(ch.isupper()):
        i = ord(ch) - 65
    else:
        i = ord(ch) - 97 + 26
    #print(i, end = ' ')
    if(i == 0):
        return '000000'
    while(i != 0):
        bit.append(abs(i%2))
        i = int(i/2)
    length = len(bit)
    bit_str = ''.join(str(i)for i in bit)[::-1]
    while (length < 6):
        bit_str = '0' + bit_str 
        length += 1
    if(len(bit_str) != 6):
        print("ALERT")
    #print(ch, bit_str)
    return bit_str
```

```python
def tobinary_ASCII(ch):
    bit = []
    i = ord(ch)
    if(i > 256):
        return '00000000'
    #print(i, end = ' ')
    if(i == 0):
        return '00000000'
    while(i != 0):
        bit.append(abs(i%2))
        i = int(i/2)
    length = len(bit)
    bit_str = ''.join(str(i)for i in bit)[::-1]
    while (length < 8):
        bit_str = '0' + bit_str 
        length += 1
    if(len(bit_str) != 8):
        print("ALERT")
    #print(ord(ch), ch, bit_str)
    return bit_str
```

```python
def tobinary_ASCII_unwritable(i):
    bit = []
    if(i > 256):
        return '00000000'
    #print(i, end = ' ')
    if(i == 0):
        return '00000000'
    while(i != 0):
        bit.append(abs(i%2))
        i = int(i/2)
    length = len(bit)
    bit_str = ''.join(str(i)for i in bit)[::-1]
    while (length < 8):
        bit_str = '0' + bit_str 
        length += 1
    if(len(bit_str) != 8):
        print("ALERT")
    #print(i, bit_str)
    return bit_str
```

```python
def generate_bit_file_from_text(text_file_name, bit_file_name):
    bit_file = open(bit_file_name, 'w')
    text_file = open(text_file_name, 'r')
    ch = text_file.read(1)
    while(ch):
        bit_file.write(tobinary_ASCII(ch))
        ch = text_file.read(1)
    return 
```

```python
text_file_name = 'Corpuses/ASCII_256_all.txt'
bit_file_name = 'Corpuses/bit_ASCII_256_all.txt'
generate_bit_file_from_text(text_file_name, bit_file_name)

text = ''
for i in range(0, 256):
    text += tobinary_ASCII_unwritable(i)
    
bit_file = open(bit_file_name, 'a')
bit_file.write(text)
bit_file.close()
```
